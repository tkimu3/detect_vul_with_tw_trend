```python
# CVE listにflag_CR,flag_Hを追加
import pandas as pd
```

```python
df = pd.read_csv("CVEs_Over10authors_20230326_except_NonScores.csv", index_col = 0)
df
```

```python
df_temp = df.copy()
```

```python
df_temp
```

```python
def setCRflag(row):
    if row['baseSeverity'] == 'CRITICAL':
        row['flag_CR'] = 1
    else:
        row['flag_CR'] = 0
    
    return row
    
df_withCRflag = df_temp.apply(setCRflag, axis = 1)
df_withCRflag.head(10)
```

```python
def setHflag(row):
    if ((row['baseSeverity'] == 'CRITICAL') | (row['baseSeverity'] == 'HIGH')):
        row['flag_H'] = 1
    else:
        row['flag_H'] = 0
    
    return row
df_withFlags = df_withCRflag.apply(setHflag, axis = 1)
df_withFlags
```

```python
df_withFlags.to_csv("CVEs_Over10authors_20230326_except_NonScores_with_flags.csv")
```
