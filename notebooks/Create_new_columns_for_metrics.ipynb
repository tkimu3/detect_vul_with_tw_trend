### Read the CSV file of rounded time series data

```python
import pandas as pd
df = pd.read_csv("work/51420_labeled_data_rounded_20230326.csv", index_col = 1)
df.rename({'Unnamed: 0' :'num'}, inplace = True, axis = 1)
df
```

### Define the function that convert str to dict in the metrics column

```python
import json
def str_to_dict(row):
    str = row['public_metrics']
    row['public_metrics_dict'] = json.loads(str.replace('\'','\"'))
    return row
```

```python
df_dict = df.apply(str_to_dict, axis = 1)
df_dict
```

### Define the function that extract 'retweet_count' from the 'public_metrics' column in each row

```python
def extract_metrics(row):
    row['retweet_count'] = row['public_metrics_dict'].get('retweet_count', None) 
    row['reply_count'] = row['public_metrics_dict'].get('reply_count', None) 
    row['like_count'] = row['public_metrics_dict'].get('like_count', None) 
    row['quote_count'] = row['public_metrics_dict'].get('quote_count', None) 
    row['impression_count'] = row['public_metrics_dict'].get('impression_count', None) 
    return row
```

```python
df_with_metrics = df_dict.apply(extract_metrics, axis = 1)
df_with_metrics
```

```python
df_with_metrics.to_csv('51420_labeled_data_rounded_with_metrics_20230326.csv')
```
