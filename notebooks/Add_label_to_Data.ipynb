```python
df3 = df2.loc[~(na_row), :]
```

```python
df3.head()
```

```python
len(df3)
```

```python
def setCRflag(row):
    if row['baseSeverity'] == 'CRITICAL':
        row['flag_CR'] = 1
    else:
        row['flag_CR'] = 0
    
    return row
    
df4_withCRflag = df3.apply(setCRflag, axis = 1)
df4_withCRflag.head(10)
```

```python
def setHflag(row):
    if ((row['baseSeverity'] == 'CRITICAL') | (row['baseSeverity'] == 'HIGH')):
        row['flag_H'] = 1
    else:
        row['flag_H'] = 0
    
    return row
df4 = df4_withCRflag.apply(setHflag, axis = 1)
df4.head(10)
```
