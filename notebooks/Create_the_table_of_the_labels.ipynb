```python
# pd.to_datetime()
# 日時（日付・時間）を表した文字列の列pandas.Seriesをdatetime64[ns]型に変換
print(pd.to_datetime(df["created_at"]))
```

```python
df["created_at"] = pd.to_datetime(df["created_at"])
df["created_at"].dtypes
```

```python
# 時系列データの丸め込み
# https://www.deep-rain.com/programming/python/1355
# 　Series.dt.round()に渡す第一引数を”H”にして、時間に丸める
print(df["created_at"].dt.round("H"))
```

```python
# https://stephenallwright.com/pandas-datetime-round-to-hour/
# Use "floor" operation to round Pandas DateTime down to nearest hour
```

```python
print(df["created_at"].dt.floor("H"))
```
