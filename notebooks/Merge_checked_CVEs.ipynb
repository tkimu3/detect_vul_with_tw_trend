```python
import pandas as pd
df_checked = pd.read_csv("checked_reasons_for_non_severity_cves_at_20230406.csv", usecols = ['CVE_checked','reason'])
df_checked
```

```python
df_reserved = df_checked[df_checked["reason"] == "RESERVED"]
df_reserved
```

```python
df_NotReserved = df_checked[df_checked["reason"] != "RESERVED"]
df_NotReserved
```

```python
df_gotScore = df_checked[(df_checked["reason"] != "RESERVED") & (df_checked["reason"] != "CVE_ID_NOT_FOUND") & ((df_checked["reason"] != "NVD score not yet provided."))]
df_gotScore
```

```python
df_gotScore[["reason"]]
```

```python
# baseSeverity = []
# baseScore = []
# for i, row in df_gotScore.iterrows():
#     df_gotScore.loc[i, 'baseSeverity'](row['reason'].split()[-1])
#     df_gotScore.loc[i, 'baseScore'](row['reason'].split()[0])
# df_gotScore

# https://www.learning-nao.com/?p=3453
# apply()で行ごと（あるいは列ごと）に独自に定義した関数を呼び出し,新たにbaseSeverityとbaseScoreの列を追加する

def SetSeverity(row):
    row['baseSeverity'] = row['reason'].split()[-1]
    return row

def SetScore(row):
    row['baseScore'] = row['reason'].split()[0]
    return row



df_gotScore2 = df_gotScore.apply(SetSeverity, axis = 1)
df_gotScore2 = df_gotScore2.apply(SetScore, axis = 1)
df_gotScore2.drop('reason', axis = 1)
```

### Add the column of num_of_author to df_gotScore2

```python
df_CVEs10p_cvss = pd.read_csv("CVEs_Over10authors_20230326.csv", header=0)
df_CVEs10p_cvss
```

```python
df_NoneSeverityCVEs = pd.read_csv("non_severity_cves_20230326_determined_at_20230406.csv")
df_NoneSeverityCVEs.head()
```

```python
# def pickCheckedCVE(row):
#     if row[['CVE']].isin(df_gotScore2['CVE_checked'].tolist()):
#         print(row['CVE'], row['num_of_author'])

# df_NoneSeverityCVEs.apply(pickCheckedCVE, axis = 1)

df_temp = df_NoneSeverityCVEs[df_NoneSeverityCVEs['CVE'].isin(df_gotScore2['CVE_checked'])]
df_temp = df_temp[['CVE','num_of_author']]
df_temp
```

```python
df_gotScore3 = df_gotScore2.rename(columns = {'CVE_checked':'CVE'}, inplace = False)
df_gotScore3 = df_gotScore3.drop('reason', axis = 1)
df_gotScore3
```

```python
df_gotScore4 = pd.merge(df_temp, df_gotScore3, how = 'left')
df_gotScore4
```

```python
df_CVEs10p_CVSS = pd.concat([df_CVEs10p_cvss, df_gotScore4], ignore_index = True)
df_CVEs10p_CVSS.tail(10)
```

```python
df_CVEs10p_CVSS.isna().sum()
```

```python
df_CVEs10p_CVSS.dropna(inplace = True)
```

```python
df_CVEs10p_CVSS = df_CVEs10p_CVSS.sort_values(['num_of_author'], axis = 0, ascending = False, ignore_index = True)
df_CVEs10p_CVSS
```

```python
df_CVEs10p_CVSS.to_csv("CVEs_Over10authors_20230326_except_NonScores.csv", header = True)
```
