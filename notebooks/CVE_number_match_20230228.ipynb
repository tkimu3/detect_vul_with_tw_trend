```python
df_matched = df[df["text"].str.contains(r'CVE-[0-9]{4}-[0-9]{4,5}')]
df_matched.head()
```

```python
# The number of tweets matched with the CVE number form
len(df_matched)
```

```python
# Extract only the CVE from the string matched "text"

# https://note.nkmk.me/python-re-match-object-span-group/
# Python„ÅÆÊ≠£Ë¶èË°®Áèæ„É¢„Ç∏„É•„Éº„É´re„ÅÆmatch()„ÇÑsearch()„ÅØ„ÄÅÊñáÂ≠óÂàó„ÅåÊ≠£Ë¶èË°®Áèæ„Éë„Çø„Éº„É≥„Å´„Éû„ÉÉ„ÉÅ„Åó„ÅüÂ†¥Âêà„ÄÅ„Éû„ÉÉ„ÉÅ„Åó„ÅüÈÉ®ÂàÜ„Çí„Éû„ÉÉ„ÉÅ„Ç™„Éñ„Ç∏„Çß„ÇØ„Éà„Å®„Åó„Å¶Ëøî„Åô„ÄÇ
# „Éû„ÉÉ„ÉÅ„Ç™„Éñ„Ç∏„Çß„ÇØ„Éà„ÅÆgroup„É°„ÇΩ„ÉÉ„Éâ„ÇíÂÆüË°å„Åô„Çã„Åì„Å®„Åß„Éû„ÉÉ„ÉÅ„Åó„ÅüÊñáÂ≠óÂàó„ÇíÊäΩÂá∫„Åß„Åç„Çã

import re
list_CVE = []
df_plusCVE = df_matched
for i, s in enumerate(df_matched["text"].to_list()):
    matched_text = re.search('CVE-[0-9]{4}-[0-9]{4,5}',s)
    if i < 5:
        print(matched_text.group())
        # df_plusCVE = pd.concat([df_plusCVE, pd.DataFrame({"CVE": pd.Series([matched_text.group()])})], axis=1)

    list_CVE.append(matched_text.group())

df_plusCVE['CVE'] = list_CVE
 
# print(list_CVE[:5])
# print(list_CVE[-5:])
# print(len(list_CVE))
df_plusCVE.head(10)
```

```python
# https://stackoverflow.com/questions/40454030/count-and-sort-with-pandas
# https://stackoverflow.com/questions/41415017/count-unique-values-using-pandas-groupby
df_ids = df_plusCVE[["author_id", "CVE"]].groupby(["CVE"])["author_id"].nunique()
df_ids
```

```python
# https://stackoverflow.com/questions/40454030/count-and-sort-with-pandas
df_ids_sorted = df_ids.reset_index(name = 'count').sort_values(['count'], ascending = False)
df_ids_sorted
```

```python
df_over10 = df_ids_sorted[df_ids_sorted['count']>= 10]
df_over10
```
