## Create the time series data of reply_count

```python
df_like = df[:]
df_like
```

```python
df_like.columns
```

```python
df_like.drop(columns =['num', 'created_at', 'id', 'author_id',
       'edit_history_tweet_ids', 'public_metrics', 'baseSeverity',
       'baseScore', 'flag_CR', 'flag_H', 'public_metrics_dict',
       'retweet_count', 'reply_count', 'quote_count',
       'impression_count'] ,inplace = True)
df_like
```

### sums up the number of reply_count

```python
df_like_sum = pd.DataFrame(df_like.groupby(['rd_created_at','CVE'])['like_count'].sum())
df_like_sum
```

```python
# df_like_sum = df_like_sum.reset_index().set_index('rd_created_at')
df_like_sum.reset_index(level = -1, inplace = True)
df_like_sum
```

## Create the timeseries data of 48 hours from the tweets were tweeted 

```python
CVE_list = df_like_sum['CVE'].unique()
len(CVE_list)
```

```python
import datetime
import numpy as np

df_48hour = pd.DataFrame()

for i, cve in enumerate(CVE_list):
    
    # create null DF
    df_48hour_ = pd.DataFrame()
    
    # Define the start & end time (48hours) for the cve-id
    start_time = df_like_sum[df_like_sum['CVE'] == cve].index[0]
    end_time = start_time + datetime.timedelta(hours = 48)

    # Create a temporary Dataframe for the cve
    df_48hour_ = df_like_sum[df_like_sum['CVE'] == cve].loc[start_time:end_time]

    # Concatenate temporary Dataframes
    if (i == 0):
        df_48hour = df_48hour_
        
    else:
        df_48hour = pd.concat([df_48hour, df_48hour_])
df_48hour
```

```python
df_ = df_48hour.reset_index()
df_
```
